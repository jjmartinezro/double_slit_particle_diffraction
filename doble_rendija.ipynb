```python
# importing needed libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
# function to generate the intensity based on the parameters

def double_slit_diffraction_intensity(slit_width, wavelength, screen_distance, distance_between_slits,  X) :
  """
    Takes in slit_width, wavelength, screen distance, distance between the two strings and a numpy array X(an array of distances from the center).
    Outputs an array of normalized intensities corresponding to X.
  """
  return (((np.sinc((np.pi*slit_width*X)/(wavelength*screen_distance)/np.pi)))**2)*((np.cos((np.pi*distance_between_slits*X)/(wavelength*screen_distance)))**2)
```

```python
X = np.arange(-0.005,0.005,0.00001)
slit_widths = [100*(10**-6), 150*(10**-6), 200*(10**-6)]
wavelengths = [250*(10**-9), 500*(10**-9), 750*(10**-9)]
screen_distances = [30*(10**-2), 50*(10**-2), 100*(10**-2)]
distances_between_slits = [1*10**-3, 3*10**-3, 5*10**-3]
```

```python
# let's evaluate how changing the slit width affects the interference pattern
for slit_width in slit_widths:
  wavelength = wavelengths[1]
  screen_distance = screen_distances[0]
  distance_between_slits = distances_between_slits[0]
  Y = double_slit_diffraction_intensity(slit_width, wavelength, screen_distance, distance_between_slits, X)
  plot, = plt.plot(X,Y)
  plt.xlabel(f"Distancia al centro (m), anchura de rendijas: {'{:.1e}'.format(float(slit_width))}")
  plt.ylabel("Intensidad (norm.)")
  plt.savefig(f"slit_width_{'{:.1e}'.format(float(slit_width))}.png")
  plt.show()
```

```python
# let's evaluate how changing the wavelength affects the interference pattern
for wavelength in wavelengths:
  slit_width = slit_widths[0]
  screen_distance = screen_distances[0]
  distance_between_slits = distances_between_slits[0]
  Y = double_slit_diffraction_intensity(slit_width, wavelength, screen_distance, distance_between_slits, X)
  plot, = plt.plot(X,Y)
  plt.xlabel(f"Distancia al centro (m), longitud de onda: {'{:.1e}'.format(float(wavelength))}")
  plt.ylabel("Intensidad (norm.)")
  plt.savefig(f"wavelength{'{:.1e}'.format(float(wavelength))}.png")
  plt.show()
```

```python
# let's evaluate how changing the screen distance affects the interference pattern
for screen_distance in screen_distances:
  slit_width = slit_widths[0]
  wavelength = wavelengths[1]
  distance_between_slits = distances_between_slits[0]
  Y = double_slit_diffraction_intensity(slit_width, wavelength, screen_distance, distance_between_slits, X)
  plot, = plt.plot(X,Y)
  plt.xlabel(f"Distancia al centro (m), distancia a la pantalla: {'{:.1e}'.format(float(screen_distance))}")
  plt.ylabel("Intensidad (norm.)")
  plt.savefig(f"screen_distance{'{:.1e}'.format(float(screen_distance))}.png")
  plt.show()
```

```python
# let's evaluate how changing the distance between slitsa affects the interference pattern
for distance_between_slits in distances_between_slits:
  slit_width = slit_widths[0]
  wavelength = wavelengths[1]
  screen_distance = screen_distances[0]
  Y = double_slit_diffraction_intensity(slit_width, wavelength, screen_distance, distance_between_slits, X)
  plot, = plt.plot(X,Y)
  plt.xlabel(f"Distancia al centro (m), distancia entre rendijas: {'{:.1e}'.format(float(distance_between_slits))}")
  plt.ylabel("Intensidad (norm.)")
  plt.savefig(f"distance_between_slits{'{:.1e}'.format(float(distance_between_slits))}.png")
  plt.show()
```
